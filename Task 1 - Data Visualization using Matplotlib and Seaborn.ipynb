```python
#loading necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

LOAD IRIS DATASET USING SCIKIT LEARN LIBRARY

```python
#import scikit learn library
from sklearn.datasets import load_iris
#load the iris dataset
iris = load_iris()
iris
```

```python
iris.data.shape
```

```python
iris.target.shape
```

CONVERTING DATA INTO PANDAS DATAFRAME

```python
#converting data into pandas dataframe
df = pd.DataFrame(iris.data, columns = iris.feature_names)
#including the target
df['target'] = iris.target
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

```python
#full dataset description
print(iris.DESCR)
```

```python
#seeing statistical information
df.describe()
```

```python
print(df.groupby('target').size())
```

**DATA VISUALIZATION**

```python
#data visualization
#box and whisker plots
#box plot - gives information about outliers
df.plot(kind='box', subplots = True, layout = (3,2), figsize = (8,12));
```

```python
#scatterplot - to get insights from data
df.plot(kind='scatter', x = 'petal length (cm)', y = 'petal width (cm)', layout = (3,2));
```

```python
df.plot(kind='scatter', x = 'sepal length (cm)', y = 'sepal width (cm)', layout = (3,2));
```

```python
#histogram to see distribution of data
df.hist(figsize = (12,12))
plt.show()
```

```python
#correlation plots
df.corr()
```

```python
#heatmap
sns.heatmap(df.corr(), annot = True, cmap = 'bone');
```

```python
#pairplot - to see relation among all the attributes
sns.pairplot(df);
```

```python
#categorical plot
sns.catplot(x = 'petal length (cm)', y = 'petal width (cm)', palette = 'bright', hue = 'target', data = df);
```
